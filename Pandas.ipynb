```python
data_nieuwe_namen = data_nl.copy()
data_nieuwe_namen.columns = ['FOOD_ID', 'code', 'COUNTRY', 'efsaprodcode2', 'voedselnaam',
       'level1', 'level2', 'level3', 'NUTRIENT_ID', 'Voedingstof', 'UNIT',
       'mg/100g']
data_nieuwe_namen.head()
```

```python
data_nieuwe_namen['UNIT'].unique()
```

```python
nieuwe_waarden = []

for index, row in data_nieuwe_namen.iterrows():
    if row['UNIT'].startswith('Microgram'):
        nieuwe_waarden.append(row['mg/100g']/1000)
    else:
        nieuwe_waarden.append(row['mg/100g'])
```

```python
data_aangepast = data_nieuwe_namen.copy()
data_aangepast['mg/100g'] = nieuwe_waarden
data_aangepast.head()
```

```python
data_dropped = data_aangepast.drop(columns=['FOOD_ID', 'code', 'COUNTRY', 'efsaprodcode2',
       'level1', 'level2', 'level3', 'NUTRIENT_ID', 'UNIT'])
data_dropped.head()
```

```python
len(data_dropped)
```

```python
data_pivoted = data_dropped.pivot_table(index='voedselnaam', columns='Voedingstof', values='mg/100g')
data_pivoted.head()
```

```python
data_pivoted.columns
```

```python
data_pivoted.columns.name = 'mg/100g'
data_pivoted.head()
```

# visualisatie

```python
plt.plot(range(10,20))
plt.show()
```

```python
plt.plot(range(len(data_pivoted)), data_pivoted['Calcium (Ca)'])
plt.show()
```

```python
plt.plot(range(len(data_pivoted)), data_pivoted['Calcium (Ca)'])
plt.ylabel('Calcium Ca(mg/100g)')
plt.xlabel('Index van voedingsmiddelen')
plt.title ('Hoeveelheid calcium van verschillende voedingsmiddelen')
plt.show()
```

```python
plt.hist(data_pivoted['Calcium (Ca)'])
plt.title ('Hoeveelheid calcium van verschillende voedingsmiddelen')
plt.ylabel('Aantal voedingsmiddelen')
plt.xlabel('Ca (mg/100g)')
plt.show()
```

```python
plt.hist(data_pivoted['Calcium (Ca)'], log=True, bins=50)
plt.title ('Hoeveelheid calcium van verschillende voedingsmiddelen')
plt.ylabel('Aantal voedingsmiddelen')
plt.xlabel('Ca (mg/100g)')
plt.show()
```

```python
plt.bar(data_pivoted.columns, data_pivoted.mean())
plt.ylabel('Hoeveelheid mg/100g')
plt.xlabel('Voedingstof')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.barh(data_pivoted.columns, data_pivoted.mean())
plt.xlabel('Hoeveelheid mg/100g')
plt.ylabel('Voedingstof')
plt.show()
```
